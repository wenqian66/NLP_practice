```python
from sec_api import ExtractorApi
import os
import sec_API_KEY#the .py file that store my api key
api_key = sec_API_KEY.API_KEY
extractorApi = ExtractorApi(api_key)

# Tesla 10-K filing
filing_url = "https://www.sec.gov/Archives/edgar/data/1318605/000156459021004599/tsla-10k_20201231.htm"

# get the standardized and cleaned text of section 1A "Risk Factors"
section_text = extractorApi.get_section(filing_url, "1A", "text")

print(section_text)
```

```python
# get the original HTML of section 7 "Management’s Discussion and Analysis of Financial Condition and Results of Operations"
section_html = extractorApi.get_section(filing_url, "7", "html")
print(section_html)
```

```python
#load the file that contain all the urls links for the 10-k that filed in 2024
file_path = 'filing_urls.txt'  # Update this with the path to your text file
output_folder = '10k_risk_factors'  # Directory to store the output files


os.makedirs(output_folder, exist_ok=True)

# Read URLs from the file
with open(file_path, 'r') as file:
    urls = file.readlines()

# Loop through each URL, extract the "Risk Factors" section, and save it to a text file
for url in urls:
    url = url.strip()  # Remove any leading/trailing whitespace
    if not url:
        continue
    
    # Extract "Risk Factors" section
    try:
        section_text = extractorApi.get_section(url, "1A", "text")
        
        # Determine a filename based on the URL or another unique identifier
        filename = os.path.join(output_folder, f"{url.split('/')[-1].split('.')[0]}_risk_factors.txt")
        
        # Write the extracted text to a file
        with open(filename, 'w') as output_file:
            output_file.write(section_text)
        print(f"Processed {filename}")
        
    except Exception as e:
        print(f"Failed to process {url}: {e}")

# Output message when complete
print("Extraction complete.")
```
